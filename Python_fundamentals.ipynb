```python
# Let's see this in action:
text = input("Enter a number: ")  # User enters: 123
print(f"String: {text}")  # "123"
print(f"ASCII values: {[ord(c) for c in text]}")  # [49, 50, 51]

num = int(text)
print(f"Integer: {num}")  # 123
print(f"Binary: {bin(num)}")  # 0b1111011

# Output:
# Enter a number: 123
# String: 123
# ASCII values: [49, 50, 51]
# Integer: 123
# Binary: 0b1111011
```

```python
# Memory limit: 2^16 = 65,536 bytes (64KB)
max_memory_16bit = 2**16  # 65,536 bytes
print(f"16-bit max RAM: {max_memory_16bit/1024}KB")

# Example: Intel 8086, DOS, early Windows
```

```python
# Memory limit: 2^32 = 4,294,967,296 bytes (4GB)
max_memory_32bit = 2**32  # 4,294,967,296 bytes
print(f"32-bit max RAM: {max_memory_32bit/1024**3}GB")

# Example: Pentium processors, Windows 95/XP
```

```python
# Memory limit: 2^64 = 18.4 million TB (theoretically)
max_memory_64bit = 2**64  # 18,446,744,073,709,551,616 bytes
print(f"64-bit max RAM: {max_memory_64bit/1024**4}TB theoretically")

# But practical limits are lower due to hardware constraints
practical_limit = 2**48  # Modern CPUs use 48-bit addressing
print(f"Practical limit: {practical_limit/1024**4}TB")
```

```python
# Write text
with open("read.txt", "w") as f:
    f.write("First line\n")
    f.write("Second line\n")

# Write multiple lines
lines = ["Line 1", "Line 2", "Line 3"]
with open("read.txt", "a") as f:
    f.writelines(line + "\n" for line in lines)

# Append to existing file
with open("read.txt", "a") as f:
    f.write("This line is appended\n")
```

```python
def sum_all(*numbers):
    total=0
    for num in numbers:
        total += num
    return total

print(sum_all(1,2,3,4))   
```

```python
def print_args(*args):
    for i,arg in enumerate(args):
        print(f"Argument: {i} {arg}")

print_args("apple", 42, True, [1, 2])        
```

```python
def get_stats(numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    average = sum(numbers) / len(numbers)
    return minimum, maximum, average   # Return Tuple

stats = get_stats([1,2,3,4,5])

min_val, max_val, avg_val = stats  # Tuple unpacking

def get_user():
    return {"name": "Vishal","age":30, "city": "New Work"}

print(min_val)
print(max_val)
print(avg_val)
print(get_user())
```
